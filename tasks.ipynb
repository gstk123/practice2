```python
# 1) Load a dataset using Pandas and perform data cleaning operations such as removing null values, removing duplicates,number format etc.
```

```python
data=pandas.read_csv(r"C:\Users\gstak\Downloads\archive (2)\all_csv sorted.csv")
data
```

```python
data.dropna(subset=["NO. OF Internet Plans"], inplace = True)
data
```

```python
data["NO. OF Internet Plans"] = data["NO. OF Internet Plans"].astype(int)
data
```

```python
# 2) Use Pandas to calculate summary statistics for a dataset, such as mean, median, and standard deviation.
```

```python
data["Average price of 1GB (USD)"] = data["Average price of 1GB (USD)"].astype(float)
price_mean = data["Average price of 1GB (USD)"].mean()
print(price_mean)
```

```python
data["Average price of 1GB (USD)"] = data["Average price of 1GB (USD)"].astype(float)
price_median = data["Average price of 1GB (USD)"].median()
print(price_median)
```

```python
data["Average price of 1GB (USD)"] = data["Average price of 1GB (USD)"].astype(float)
price_std = data["Average price of 1GB (USD)"].std()
print(price_std)
```
